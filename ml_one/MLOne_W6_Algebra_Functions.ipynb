# Sigmoid ðŸ¤©

Here is the equation of sigmoid function:

$Sigmoid(x) = \frac{1}{1 + e^{-x}} $

````python
def sigmoid(x):
	return 1.0/(1.0+np.exp(-x))
````

Similar to above, we can plot the function graph between the range -10 and 10:

````python
x = np.arange(-10.0, 10.0, 0.5)
y = sigmoid(x)
plt.plot(x,y)
plt.show()
````

It may seem that some output values have reached the boundaries of 0 and 1, however if you look into the actual values of y:

````python
print(y)
````

It just gets reeeally close to 0 and 1 but not exactly 0 and 1.

# Relu ðŸ¤©

The [relu](https://en.wikipedia.org/wiki/Rectifier_(neural_networks)) function is defined as follows

$\text{relu}(x) = \left\{
   \begin{array}{rl}
     x, & \text{if }  x \ge 0 \\
     0, & \text{otherwise }.
   \end{array}\right.$

There are several ways to implement.

````python
# First way: with help from np.maximum()
def Relu_FirstImpementation(x):
  return np.maximum(0,x)
````

````python
x = np.arange(-5.0, 5.0, 0.5)
y = Relu_FirstImpementation(x)
plt.plot(x,y)
plt.show()
````

````python
# Second way:
# noodling time: how does this work?
def Relu_SecondImpementation(x):
  return x*(x>0)
````

because (x>0) returns 1 if it is true and 0 if it is false



````python
x = np.arange(-5.0, 5.0, 0.5)
y = Relu_SecondImpementation(x)
plt.plot(x,y)
plt.show()
````
