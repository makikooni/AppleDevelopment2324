```python
# predicting what will happen

import random

x = random.random()

print(x)

if x < 0.2:
  print("apples")
elif x < 0.6:
  print("oranges")
elif x >= 0.6:
  print("pears")
```

```python
# Simple Markov Process
import random
import time

def markov(state):
  x = random.random()
  if state == "apples":
    if x<0.4:
      state = "oranges"
    elif x<0.6:
      state = "pears"
    elif x>0.6:
      state = "apples"
  elif state == "oranges":
    if x<0.2:
      state = "oranges"
    elif x<0.8:
      state = "pears"
    elif x>0.8:
      state = "apples"
  elif state == "pears":
    if x<0.7:
      state = "oranges"
    elif x<0.9:
      state = "pears"
    elif x>0.9:
      state = "apples"
  print(x,state)
  time.sleep(1)
  markov(state)

while True:
  try :
    markov("apples")
  except KeyboardInterrupt:
    break
```

```python
# Simple Markov Process
import random
import time

states = []

def markov(state):
  x = random.random()
  if state == "Beer":
    if x<0.4:
      state = "Cider"
    elif x<0.6:
      state = "Vodka"
    elif x>0.6:
      state = "Beer"
  elif state == "Cider":
    if x<0.2:
      state = "Cider"
    elif x<0.8:
      state = "Vodka"
    elif x>0.8:
      state = "Beer"
  elif state == "Vodka":
    if x<0.7:
      state = "Cider"
    elif x<0.9:
      state = "Beer"
    elif x>0.9:
      state = "Vodka"
  states.append(state)
  print(x,state)
  time.sleep(0.1)
  markov(state)

while True:
  try :
    markov("Beer")
  except KeyboardInterrupt:
    break
```

```python
print(states.count("Beer"))
print(states.count("Cider"))
print(states.count("Vodka"))
```
